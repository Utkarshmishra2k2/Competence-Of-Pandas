```python
data_01.Name.fillna("Unknown",inplace = True)
data_01.Gender.ffill(inplace = True)
data_01.Management.fillna(data_01.Management.mode()[0],inplace = True)
data_01.Team.fillna("Independent",inplace = True)
```

```python
data_01.isnull().sum()
```

```python
data_01.duplicated().sum()
```

```python
stats.iqr(data_01.Salary, interpolation = 'midpoint')
```

```python
data_02 = data_01[data_01.Salary.between(np.percentile(data_01.Salary, 25, interpolation = 'midpoint'), np.percentile(data_01.Salary, 75, interpolation = 'midpoint'))]
data_02 = data_01[data_01.Bonus.between(np.percentile(data_01.Bonus, 25, interpolation = 'midpoint'), np.percentile(data_01.Bonus, 75, interpolation = 'midpoint'))]
```

```python
data_02["BonusRupess"] = ((data_02.Salary)*(data_02.Bonus))/100
```

```python
data_02
```

```python
px.scatter(data_02, x="Salary", y="BonusRupess", color="Team",size="Bonus",title="Bonus Rupess VS Salary VS Team VS Bonus ",
          labels={ "Salary":"Salary","BonusRupess":"Bonus Rupess"
                 })
```

```python
px.line(data_02, x="Salary", y="BonusRupess", color="Team",title="Bonus Rupess VS Salary VS Team ",
          labels={ "Salary":"Salary","BonusRupess":"Bonus Rupess"
                 })
```

```python
px.histogram(data_02, x="Salary",title="Bonus Rupess VS Salary VS Team VS Bonus ",
          labels={ "Salary":"Salary","BonusRupess":"Bonus Rupess"
                 }).update_layout(bargap=0.2)
```

```python
px.pie(data_02, values='Bonus', names='Team', title="Team VS Bonus")
```

```python
px.area(data_02, x="Salary", y="BonusRupess", color="Team",title="Bonus Rupess VS Salary VS Team VS Bonus ",
          labels={ "Salary":"Salary","BonusRupess":"Bonus Rupess"
                 })
```
